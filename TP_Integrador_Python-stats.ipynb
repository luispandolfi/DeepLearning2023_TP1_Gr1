## Estadísticas

Las estadísticas recomendadas para cada tipo de datos son las siguientes, pueden agregar más si lo desean:

```python
import matplotlib.pyplot as plt
```

# Estadísticas Persona

## Totales

```python
stats_personas = Persona.get_stats(df_personas)
print(f'El total de personas es: {stats_personas["total_personas"]}')
print(f'La persona más vieja es: {stats_personas["persona_mas_vieja"]}')
print(f'La persona más joven es: {stats_personas["persona_mas_joven"]}')

print()
stats_personas = Persona.get_stats(df_personas, genero='F')
print(f'El total de mujeres es: {stats_personas["total_personas"]}')

print()
stats_personas = Persona.get_stats(df_personas, genero='M')
print(f'El total de hombres es: {stats_personas["total_personas"]}')

print()
stats_personas = Persona.get_stats(df_personas, fecha_nacimiento=[1980, 1983])
print(f'El total de personas nacidas entre 1980 y 1983 es: {stats_personas["total_personas"]}')
```

## Personas por año de nacimiento

```python
stats_personas = Persona.get_stats(df_personas)
personas_por_anio = stats_personas["personas_por_anio"]
plt.bar(personas_por_anio.index, personas_por_anio.values)
plt.show()
```

## Personas por género

```python
stats_personas = Persona.get_stats(df_personas)
personas_por_genero = stats_personas["personas_por_genero"]
plt.bar(personas_por_genero.index, personas_por_genero.values)
plt.show()
```

# Estadísticas de Usuario

## Totales

```python
stats_usuarios = Usuario.get_stats(df_usuarios)
print(f'El total de usuarios es: {stats_usuarios["total_usuarios"]}')

print()
stats_usuarios = Usuario.get_stats(df_usuarios, ocupacion='scientist')
print(f'El total de usuarios con ocupacion "scientist" es: {stats_usuarios["total_usuarios"]}')

print()
stats_usuarios = Usuario.get_stats(df_usuarios, ocupacion='scientist', fecha_alta=['1997-01-01', '1997-12-31'])
print(f'El total de usuarios con ocupacion "scientist" dados de alta en 1997 es: {stats_usuarios["total_usuarios"]}')
```

## Usuarios por fecha de alta

```python
stats_usuarios = Usuario.get_stats(df_usuarios)
usuarios_por_anio = stats_usuarios["usuarios_por_anio"]

plt.bar(usuarios_por_anio.index, usuarios_por_anio.values)
# configurar eje x para que muestre solo valores enteros
ax = plt.gca()
ax.xaxis.set_major_locator(plt.MultipleLocator(base=1))
plt.show()
```

## Usuarios por ocupación

```python
stats_usuarios = Usuario.get_stats(df_usuarios)
usuarios_por_ocupacion = stats_usuarios["usuarios_por_ocupacion"]
plt.barh(usuarios_por_ocupacion.index, usuarios_por_ocupacion.values)
plt.show()
```
